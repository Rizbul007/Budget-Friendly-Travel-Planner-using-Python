```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample destinations and costs
data = {
    "Destination": ["Chittagong", "Dhaka", "Sylhet", "Pabna", "Cox's Bazar"],
    "Transportation Cost (Tk)": [7000, 5000, 9000, 3000, 8000],
    "Accommodation Cost per Day (Tk)": [1200, 1500, 1100, 500, 1300],
    "Food Cost per Day (Tk)": [500, 700, 400, 200, 600]
}

destinations = pd.DataFrame(data)
destinations
```

```python
def get_travel_details():
    print("Budget-Friendly Travel Planner")
    budget = float(input("Enter your total budget (in Tk): "))
    days = int(input("Enter the number of days for the trip: "))
    return budget, days
```

```python
def calculate_costs(destinations, days):
    destinations["Total Cost (Tk)"] = (
        destinations["Transportation Cost (Tk)"] +
        destinations["Accommodation Cost per Day (Tk)"] * days +
        destinations["Food Cost per Day (Tk)"] * days
    )
    return destinations
```

```python
def suggest_destinations(destinations, budget):
    affordable_destinations = destinations[destinations["Total Cost (Tk)"] <= budget]
    if affordable_destinations.empty:
        print("No destinations available within your budget.")
    else:
        print("\nDestinations within your budget:")
        print(affordable_destinations[["Destination", "Total Cost (Tk)"]])
```

```python
def visualize_expenses(destinations, destination_name, days):
    destination = destinations[destinations["Destination"] == destination_name]
    if destination.empty:
        print("Destination not found.")
        return

    transport = destination["Transportation Cost (Tk)"].values[0]
    accommodation = destination["Accommodation Cost per Day (Tk)"].values[0] * days
    food = destination["Food Cost per Day (Tk)"].values[0] * days

    expenses = [transport, accommodation, food]
    labels = ["Transportation", "Accommodation", "Food"]

    plt.figure(figsize=(8, 6))
    plt.pie(expenses, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'lightgreen'])
    plt.title(f"Expense Breakdown for {destination_name}")
    plt.show()
```

```python
def main():
    # Get user inputs
    budget, days = get_travel_details()

    # Calculate costs
    destinations_with_costs = calculate_costs(destinations, days)

    # Suggest destinations
    suggest_destinations(destinations_with_costs, budget)

    # Ask user to visualize expenses
    choice = input("\nDo you want to see the expense breakdown for a destination? (yes/no): ").strip().lower()
    if choice == 'yes':
        destination_name = input("Enter the destination name: ").strip()
        visualize_expenses(destinations_with_costs, destination_name, days)

# Run the program
main()
```
